```python
from sklearn.datasets import fetch_openml
import pandas as pd
```

```python
# Descargar la base de datos Titanic desde OpenML
titanic = fetch_openml(name='titanic', version=1, as_frame=True)

# Convertir a DataFrame de pandas
df_titanic = titanic.frame

# Descripción básica del dataset
descripcion = df_titanic.describe(include='all')

# Mostrar el DataFrame
df_titanic.head(), descripcion
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma para la variable 'age'
plt.figure(figsize=(10, 5))
sns.histplot(df_titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribución de la Edad de los Pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

```

```python
# Gráfico de barras para la variable 'sex'
plt.figure(figsize=(6, 4))
sns.countplot(data=df_titanic, x='sex', palette='Set2')
plt.title('Distribución por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Frecuencia')
plt.show()


```

```python
# Boxplot para la variable 'fare'
plt.figure(figsize=(10, 5))
sns.boxplot(x='pclass', y='fare', data=df_titanic, palette='Set3')
plt.title('Distribución de Tarifa por Clase de Pasajero')
plt.xlabel('Clase de Pasajero')
plt.ylabel('Tarifa')
plt.show()

```

```python
# Gráfico de torta (pie chart) para la variable 'embarked'
plt.figure(figsize=(6, 6))
df_titanic['embarked'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Porcentaje de Embarques por Puerto')
plt.ylabel('')
plt.show()
```

# Ejercicios

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Gráfica de Barras (Distribución de la variable 'sex')

plt.figure(figsize=(6, 4))
sns.countplot(data=df_titanic, x='sex', palette='Set2')
plt.title('Distribución por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Frecuencia')
plt.show()
```

```python
## Gráfica de Sectores (Pastel) - Puerto de embarque

plt.figure(figsize=(6, 6))
df_titanic['embarked'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Porcentaje de Embarques por Puerto')
plt.ylabel('')
plt.show()
```

```python
## Gráfica de Dispersión (Edad vs. Tarifa)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', data=df_titanic, hue='pclass', palette='Set1')
plt.title('Dispersión de Edad vs Tarifa por Clase')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.show()
```

```python
## Histograma (Distribución de la Edad)

plt.figure(figsize=(10, 5))
sns.histplot(df_titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribución de la Edad de los Pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
## Gráfica de Líneas (Promedio de la tarifa por clase de pasajero)

fare_mean = df_titanic.groupby('pclass')['fare'].mean()

plt.figure(figsize=(8, 5))
plt.plot(fare_mean.index, fare_mean.values, marker='o', color='blue')
plt.title('Promedio de Tarifa por Clase de Pasajero')
plt.xlabel('Clase de Pasajero')
plt.ylabel('Tarifa Promedio')
plt.grid(True)
plt.show()
```

```python
## Gráfica Box-Plot (Distribución de tarifas por clase)

plt.figure(figsize=(10, 5))
sns.boxplot(x='pclass', y='fare', data=df_titanic, palette='Set3')
plt.title('Distribución de Tarifas por Clase de Pasajero')
plt.xlabel('Clase de Pasajero')
plt.ylabel('Tarifa')
plt.show()
```
